```python
import os
from google.colab import files
import shutil

download_dir = '/content/drive/My Drive/VHSRP/research/generated_images'
#download_dir = generated_images_dir_test
zip_dir = '/content/generated_image_zips' # Directory to store zip files

# Create directory for zip files if it doesn't exist
os.makedirs(zip_dir, exist_ok=True)

if os.path.exists(download_dir):
    print(f"Processing directories in: {download_dir}")
    for category_folder in os.listdir(download_dir):
        category_path = os.path.join(download_dir, category_folder)
        if os.path.isdir(category_path):
            print(f"  Compressing category: {category_folder}")
            zip_filename = f"{category_folder}.zip"
            zip_filepath = os.path.join(zip_dir, zip_filename)
            try:
                shutil.make_archive(zip_filepath.replace('.zip', ''), 'zip', category_path)
                print(f"  Compressed to: {zip_filepath}")
            except Exception as e:
                print(f"  Error compressing {category_folder}: {e}")

    print(f"\nDownloading zip files from: {zip_dir}")
    for zip_file in os.listdir(zip_dir):
        zip_file_path = os.path.join(zip_dir, zip_file)
        if os.path.isfile(zip_file_path) and zip_file.endswith(".zip"): # Modified condition to download all .zip files
            try:
                files.download(zip_file_path)
                print(f"    Downloaded: {zip_file}")
            except Exception as e:
                print(f"    Error downloading {zip_file}: {e}")
else:
    print(f"Download directory not found: {download_dir}")
```
