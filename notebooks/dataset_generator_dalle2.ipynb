```python
from openai import OpenAI
from PIL import Image
from io import BytesIO
import requests
import os

generated_images_dir = '/content/gdrive/My Drive/VHSRP/final/generated_images'
#generated_images_dir_test = '/content/gdrive/My Drive/VHSRP/final/generated_images1'

client = OpenAI(
  api_key="" # Add your actual API key
)


image_model = "dall-e-2"
num_of_images = 6 # Changed to generate 6 images per prompt
img_size= "256x256"
img_quality="standard"

#for prompts_dict in [prompts_dict_test]:
for prompts_dict in [prompts_dict_occupation, prompts_dict_activity]:
  for category, prompts in prompts_dict.items():
      print(f"Processing category: {category}")
      category_save_dir = f'{generated_images_dir}/{category}'
      os.makedirs(category_save_dir, exist_ok=True) # Ensure category directory exists

      for prompt_id, prompt in prompts:
          print(f"  Generating images for prompt: {prompt}")

          # Create a subdirectory for each prompt using prompt_id
          prompt_save_dir = os.path.join(category_save_dir, str(prompt_id))
          os.makedirs(prompt_save_dir, exist_ok=True)

          for i in range(num_of_images): # Need to add this loop as DALL-E 3 does not respect the n parameter
              try:
                  response = client.images.generate(
                    model=image_model,
                    prompt=prompt,
                    size=img_size,
                    n=1, # Generate one image at a time
                  )

                  # Access the image URL and download the image
                  image_url = response.data[0].url
                  image_response = requests.get(image_url)
                  img = Image.open(BytesIO(image_response.content))

                  # Create a safe filename from the prompt and include an index for multiple images
                  safe_prompt = "".join([c for c in prompt if c.isalnum() or c in (' ', '.', '_')]).rstrip()
                  image_filename = f"{prompt_id}_{i+1}_{safe_prompt[:30]}.png" # Use prompt_id, index and truncated safe prompt as filename
                  save_path = os.path.join(prompt_save_dir, image_filename)

                  img.save(save_path)
                  print(f"    Image {i+1} saved to: {save_path}")

              except Exception as e:
                  print(f"  Error generating image {i+1} for prompt '{prompt}': {e}")
```

# Download generated data

```python
import os
from google.colab import files
import shutil

download_dir = generated_images_dir
#download_dir = generated_images_dir_test
zip_dir = '/content/generated_image_zips' # Directory to store zip files

# Create directory for zip files if it doesn't exist
os.makedirs(zip_dir, exist_ok=True)

if os.path.exists(download_dir):
    print(f"Processing directories in: {download_dir}")
    for category_folder in os.listdir(download_dir):
        category_path = os.path.join(download_dir, category_folder)
        if os.path.isdir(category_path):
            print(f"  Compressing category: {category_folder}")
            zip_filename = f"{category_folder}.zip"
            zip_filepath = os.path.join(zip_dir, zip_filename)
            try:
                shutil.make_archive(zip_filepath.replace('.zip', ''), 'zip', category_path)
                print(f"  Compressed to: {zip_filepath}")
            except Exception as e:
                print(f"  Error compressing {category_folder}: {e}")

    print(f"\nDownloading zip files from: {zip_dir}")
    for zip_file in os.listdir(zip_dir):
        zip_file_path = os.path.join(zip_dir, zip_file)
        if os.path.isfile(zip_file_path) and zip_file.endswith("Labor.zip"):
            try:
                files.download(zip_file_path)
                print(f"    Downloaded: {zip_file}")
            except Exception as e:
                print(f"    Error downloading {zip_file}: {e}")
else:
    print(f"Download directory not found: {download_dir}")
```
